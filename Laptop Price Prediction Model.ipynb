```python
def fetch_processor(string):
        if string == 'Intel Core i5' or string == 'Intel Core i7' or string == 'Intel Core i3':
            return string
        elif 'Intel' in string:
            return 'Other Intel Processor'
        else:
            return 'AMD Processors'
```

```python
def memory_type(string):
    if '+' in  string:
        return (string.split('+')[0].split(),string.split('+')[1].split())
    else:
        return string.split()
```

```python
df['Memory'] = df['Memory'].astype(str)
index = 0
for i in df['Memory']:
    df['Memory'][index] = memory_type(i)
    index+=1
```

```python
df['Flash Storage'] = df['Memory'].apply(lambda x : x[0][0] if (type(x) == tuple and x[0][1] == 'Flash') else x[0] if (x[1] == 'Flash') else 0)
```

```python
df['HDD'] = df['Memory'].apply(lambda x : x[1][0] if (type(x) == tuple and (x[0][1] == 'HDD' or x[1][1]=='HDD')) else x[0] if (x[1] == 'HDD') else 0)
```

```python
df['SSD'] = df['Memory'].apply(lambda x : x[0][0] if (type(x) == tuple and (x[0][1] == 'SSD' or x[1][1]=='SSD')) else x[0] if (x[1] == 'SSD') else 0)
```

```python
df['Hybrid'] = df['Memory'].apply(lambda x : x[1][0] if (type(x) == tuple and (x[0][1] == 'Hybrid' or x[1][1]=='Hybrid')) else x[0] if (x[1] == 'Hybrid') else 0)
```

```python
index = 0
for x in df['Memory']:
    if (type(x)==tuple):
        if (x[0][1]=='SSD' and x[1][1]=='SSD'):
            df['SSD'][index] = (int(x[0][0].split('G')[0]) + int(x[1][0].split('G')[0]))
    index+=1
```

```python
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,3,8,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
